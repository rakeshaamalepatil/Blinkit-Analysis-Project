### **Data cleaning code**
#### This cell is all about making the data uniform. We are using the .replace() function to fix inconsistencies in the Item Fat Content column. By replacing 'LF', 'low fat', and 'reg' with 'Low Fat' and 'Regular', we ensure that all similar entries are counted together. This is a very common task in data preparation to ensure that your analysis is accurate.

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                          'low fat': 'Low Fat',
                                                          'reg': 'Regular'
                                                          })
```

### **Data cleaned**

### **Business Requirements**


### **KPI's Requirements**
#### This cell performs several important calculations on the data. It first calculates the total sales, average sales, number of items sold, and the average rating by using functions like .sum(), .mean(), and .count(). The code then uses the print() function to display these results in a clear way. The f before the string lets us easily put our calculated values directly into the text, and the parts like ,.0f or ,.1f are formatting instructions to make the numbers look nice (e.g., adding commas and setting the number of decimal places).

```python
# Total Sales
total_sales = df['Sales'].sum()

# Average Sales 
avg_sales = df['Sales'].mean()

# No of Items sold
no_of_items_sold = df['Sales'].count()

# Average of Rating
avg_rating = df['Rating'].mean()

#Display (f' added because we want the expression in the statement, and 1f is mentioned for decimal figure)
print(f"Total Sales: ${total_sales: ,.0f}")
print(f"Average Sales: ${avg_sales: ,.0f}")
print(f"No of Items sold: {no_of_items_sold: ,.0f}")
print(f"Average of Rating: {avg_rating: ,.1f}")
```

### **Chart Requirements**


#### **Total Sales by Fat Content**
#### This cell is all about visualizing the data. The first line groups the DataFrame by Item Fat Content and sums up the sales for each group, which is a good way to get a total for each category. The plt.pie() function then takes this summarized data to generate a pie chart. The chart is titled, and the plt.axis('equal') command makes sure the chart is a perfect circle, which is important for a correct visualization. The final chart shows the proportion of sales for each fat content category.

```python
sales_by_fat = df.groupby("Item Fat Content")['Sales'].sum()

plt.pie(sales_by_fat, labels= sales_by_fat.index,
                        autopct = '%.1f',
                        startangle = 90)
plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show
```

#### **Total Sales by Item Type**
#### This cell is for creating a bar chart to visualize sales performance by item type. The first line groups the DataFrame by Item Type, calculates the sum of Sales for each type, and then sorts the results in descending order. This gives us the top-selling items first. The code then uses matplotlib to generate the bar chart. It sets the title and axis labels to make the chart clear. A for loop is used to add text labels on each bar, showing the precise sales figures. This is an effective way to communicate which product categories are most profitable.

```python
sales_by_type = df.groupby("Item Type")['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6)) #size in inches
bars = plt.bar(sales_by_type.index, sales_by_type.values) #for labels 

plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars :
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
            f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize = 8)
plt.tight_layout()
plt.show()

```

#### **Fat Content by Outlet for Total Sales**
#### This cell creates a grouped bar chart to visualize sales. The code first groups the df DataFrame by both Outlet Location Type and Item Fat Content to calculate the sum of Sales for each unique combination. The .unstack() command reshapes the data so that it's easy to plot. Then, the grouped.plot(kind='bar') function creates the actual chart. The remaining lines add a title, label the axes, and add a legend to ensure the chart is easy to understand. This visualization is particularly helpful for understanding how the sales of different products vary across different types of store locations.

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8,5), title="Oultet Tier by Item Fat Content")
plt.xlabel('Outlet Loaction Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment**
#### This code helps us analyze sales trends over time. The first line groups the data by the year each outlet was established and then calculates the total sales for each year. It then generates a line plot to visualize these sales figures, with a marker on each data point. The plot is customized with labels, a title, and a for loop that adds the exact sales numbers to each point on the line. This chart is a great way to see if there is a relationship between the age of the outlets and their total sales.

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Size**
#### This code helps us understand which size of outlet contributes most to total sales. The first line groups the data by Outlet Size (e.g., small, medium, large) and calculates the total sales for each group. It then creates a pie chart to visually represent these sales. The code sets the labels and displays the percentage of sales for each slice. This makes it easy to see the sales distribution across different sizes of outlets.

```python
    sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Location - Pie Chart**
#### This code is used to create a pie chart that breaks down total sales by the type of outlet location (e.g., Tier 1, Tier 2, Tier 3). The first line of code finds the total sales for each location type. The rest of the code then takes this information and draws a pie chart, with a title and labels that show the percentage of sales for each location type. This chart is a great way to visually see which types of locations are the most profitable.

```python
sales_by_size = df.groupby('Outlet Location Type')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Location TYpe')
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Location - Bar Chart**
#### This code is for creating a bar chart to compare sales across different types of outlet locations. The first two lines of code calculate the total sales for each location type and then sort them from highest to lowest. It then uses the sns.barplot() function from the seaborn library to plot this data, creating a horizontal bar chart where the length of each bar represents the total sales. The code adds a title and labels for the axes to make the chart easy to read, allowing you to quickly see which location types are the most profitable.

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8, 3))  # Smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()  # Ensures layout fits without scroll
plt.show()
```
